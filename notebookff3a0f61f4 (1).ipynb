```python
train_data=pd.read_csv("/kaggle/input/amex-default-prediction/train_data.csv", nrows= 200000)
train_labels=pd.read_csv("/kaggle/input/amex-default-prediction/train_labels.csv", nrows=200000)
```

```python
j=0
drop_col=[]
for i in amex.columns:
    if (amex[i].isnull().sum()/len(amex[i])*100)>=25:
        drop_col.append(i)
        print("column which as >=25 of missing data:",i)
        j=j+1
print(drop_col)
print( "total number of columns:",j)
```

```python
amex.drop(drop_col,inplace=True,axis=1)
```
